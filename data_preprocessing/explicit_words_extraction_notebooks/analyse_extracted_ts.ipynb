```python
found_curse_words = pd.read_csv("D:\projects\ProfanityFilter\explicits\csvs\explicits_uris_final.csv", sep=";")
```

```python
found_curse_words = found_curse_words[~found_curse_words.content.isna()]
found_curse_words.reset_index(drop=True, inplace=True)
```

```python
final_dir = "D:\projects\ProfanityFilter\data\explicit_phrases"
for i, row in tqdm(found_curse_words.iterrows()):
    uri = row.uri.rsplit(":", 1)[-1]
    fr_id = row.fr_id
    tags = row.content
    p = os.path.join(ts_dir, f"{uri}_{fr_id}.wav")
    for tag in tags.split(" "): 
        target_dir = os.path.join(final_dir, tag)
        os.makedirs(target_dir, exist_ok=True)
        target_path = os.path.join(target_dir, f"{uri}_{fr_id}.wav")
        shutil.copy(p, target_path)
```
