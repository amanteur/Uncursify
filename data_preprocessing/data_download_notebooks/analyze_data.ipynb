```python
import os
import matplotlib.pyplot
import pandas as pd
import json
from glob import glob
from tqdm import tqdm
import requests
import subprocess
import yt_dlp
tqdm.pandas()
```

```python
csv_all = "D:\projects\ProfanityFilter\explicits\songs_info.txt"
df_all = pd.read_csv(csv_all, names=["index", "uri", "explicit", "popularity", "isrc", "artist_uri"])
df_all.set_index('index', inplace=True)
df_all.to_csv("D:\projects\ProfanityFilter\explicits\csvs\\all.csv", index=False)
```

```python
csv_explicits = "D:\projects\ProfanityFilter\explicits\songs_info_explicit.csv"
df_explicits = pd.read_csv(csv_explicits)
df_explicits.to_csv("D:\projects\ProfanityFilter\explicits\csvs\\explicits.csv", index=False)
```

```python
csv_explicits_two_versions = "D:\projects\ProfanityFilter\explicits\explicits_with_two_versions.txt"
df_explicits_two_versions = pd.read_csv(csv_explicits_two_versions, 
                                        names=["uri", "has_clean", "uri_clean", "dur", "dur_clean"])
df_explicits_two_versions.to_csv("D:\projects\ProfanityFilter\explicits\csvs\\explicits_2v.csv", index=False)
```

```python
csv_explicits_two_versions_yt_links = "D:\projects\ProfanityFilter\explicits\youtube_links_new.txt"
df_explicits_two_versions_yt_links = pd.read_csv(csv_explicits_two_versions_yt_links, 
                                                 names=["uri", "ytid", "ytid_clean", "dur_yt", "dur_yt_clean"])
df_explicits_two_versions_yt_links.to_csv("D:\projects\ProfanityFilter\explicits\csvs\\explicits_2v_yt.csv", index=False)
```

```python
df = pd.merge(df_all, df_explicits, how="left", left_on="uri", right_on="uri")
df = pd.merge(df, df_explicits_two_versions, how="left", left_on="uri", right_on="uri")
df = pd.merge(df, df_explicits_two_versions_yt_links, how="left", left_on="uri", right_on="uri")
```
