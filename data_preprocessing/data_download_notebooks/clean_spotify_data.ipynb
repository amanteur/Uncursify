```python
import os
import matplotlib.pyplot
import pandas as pd
import json
from glob import glob
from tqdm import tqdm
import requests
tqdm.pandas()
```

```python
csv_path = "D:\projects\ProfanityFilter\spotify_songs.csv"
csv_expl_path = "D:\projects\ProfanityFilter\explicits\songs_info.txt"
df_main = pd.read_csv(csv_path)
df_explicit = pd.read_csv(csv_expl_path, names=["index", "uri", "explicit", "popularity", "isrc", "artists_uri"])
```

```python
df_explicit.drop("index", axis=1, inplace=True)
```

```python
df_main.drop("explicit", axis=1, inplace=True)
```

```python
cols_to_use = df_explicit.columns.difference(df_main.columns)
```

```python
df = pd.merge(df_main, df_explicit, right_on="uri", left_on="uri", how='outer')
```

```python
df_cleaned = df[(df["explicit"]==True) & (df["popularity"]>20.0)].sort_values(by="popularity", ascending=False)
for col in df_cleaned.columns:
    if col in ["title", "artist", "isrc"]:
        df_cleaned[col] = df_cleaned[col].str.lower()
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned = df_cleaned[~df_cleaned["title"].str.contains(r'remix|live|cover')]
df_cleaned["title"] = df_cleaned["title"].replace(
    {
        r'\([^)]*\)' : '', 
        r'\[[^)]*\]': '', 
        '-': '',
        r' feat.*': '',
    }, regex=True)
```
